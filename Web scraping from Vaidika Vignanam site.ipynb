```python
from bs4 import BeautifulSoup #in cmd pip install bs4
import requests
from collections import defaultdict
from typing import Dict, List
import re
```

```python
headers = {
    "Connections":"keep-alive",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Applewebkit'537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36"
}
```

```python
# getting the html content
page =  requests.get('https://vignanam.org/index.htm#&panel1-2', headers = headers)

# we need to convert the html to text
soup = BeautifulSoup(page.content,'html.parser')

print(soup.prettify())
```

```python
languages = []

table = soup.find('table',{'class': 'deftable'})
for lang in table.findAll('a'):
    languages.append(lang.get_text())
    
print(languages)
```

```python
json_data = defaultdict(list)
link = '' #empty string

for lang in languages[:1]: #slicing[:1] is done to check if it is working for english so that same works for other langs
    if lang == 'english':
        link = 'https://vignanam.org/index.htm#&panel1-1'
        
    else:
        link = f'https://vignanam.org/{lang}.htm#&panel1-1'
        
    page = requests.get(link, headers = headers)
    soup = BeautifulSoup(page.content, 'html.parser') #html to text
    ul = soup.find('ul', {'class':'aqtree3clickable'})
    
    section_dict = dict()
    subsection_count = 0
    all_subsections = []
    
    for li in ul.findAll('a',{'class':'link4'}):
        subsection = li.get_text().replace('\xa0',' ')  #if replacement was not done the result would be Nitya\xa0parayana\xa0 - so \xa0 is replaced by space
        all_subsections.append(subsection)
        
    for li in ul.findAll('a',{'class':'link1'}):
        section = li.get_text().replace('\xa0',' ')
        count = int(re.findall(r'\d+',section)[0])
        section_dict[section] = all_subsections[subsection_count: subsection_count + count]
        subsection_count += count
```
